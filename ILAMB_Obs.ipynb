```python
import os
import xarray as xr
import numpy as np
from fates_calibration.ILAMB_obs_dict import ILAMB_DATA_DICT
```

```python
datasets = list(ILAMB_DATA_DICT.keys())
out_dir = '/glade/work/afoster/FATES_calibration/ILAMB_data'
```

```python
dataset = datasets[0]
dataset
```

```python
attributes = ILAMB_DATA_DICT[dataset]
model = attributes['model']
out_var = attributes["out_var"]
file_name = os.path.join(out_dir, f"{model}_{out_var.upper()}.nc")
ds = xr.open_dataset(file_name)
```

```python
ds_sub = ds.sel(year=slice(2003, 2014))
```
